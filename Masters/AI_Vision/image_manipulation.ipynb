# Image Manipulation
This notebook deals with simple exercises on an image. Initially, the image file was read and displayed. The size of the image and the number of color channels were then outputted. For the red channel, the maximum pixel value was identified, and for the blue channel, the pixel values at the four corners of the image were extracted and reported. The brightness level of the image was estimated by calculating the average value of the green channel. To make the image darker, specific darkness factors were applied to the red, green, and blue channels. The darkened image was displayed, and the number of channels and the average pixel value at the center point were outputted. The original image was then converted to grayscale by computing each pixel as the minimum value among the three color channels, with the original and grayscale images displayed side by side. The pixel value at the center of the grayscale image was also printed. Finally, a Keras neural network model was constructed with specific configurations. The model included a specified number of hidden layers, each with a defined size and ReLU activation, followed by dropout layers where applicable. The final layer had a specific size and sigmoid activation. The summary of the model was printed, along with the activation function names of the layers at index 1 and index 3.

```python
#Imported necessary packages as required by the instruction
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Read the file and displayed the image

dog_img = plt.imread('Dog.png')

plt.imshow(dog_img)
```

```python
#Displayed the size of the image and the number of channels

print(dog_img.shape)
```

```python
#For Channel R, displayed the maximum pixel value, which shows the highest pixel value in the channel.

redchannel = dog_img[:,:,0]

print(redchannel.max())
```

#### As expected, we had a value of 1 for the max as png files only have pixel values ranging from 0-1

```python
#For Channel B, displayed the pixel values of the four vertices (top-left, top-right, bottom-left, and bottom-right) separately

top_left_bluechannel = dog_img[0,0,2]
top_right_bluechannel = dog_img[0,1101,2]
bottom_left_bluechannel = dog_img[1119,0,2]
bottom_right_bluechannel = dog_img[1119,1101,2]

print(top_left_bluechannel)
print(top_right_bluechannel)
print(bottom_left_bluechannel)
print(bottom_right_bluechannel)
```

```python
#Got the average of the green value as this will be the brightness value 
greenchannel_average = np.average(dog_img[:,:,1])

greenchannel_average
```

```python
#Defined a function that when passed a brightness value, will display the brightness value and corresponding brightness level

def brightness(greenchannel_average):
    brightness = greenchannel_average
    if brightness < 0.2:
        return "Brightness value is " + str(brightness), "Brightness Level is Dim"
    elif brightness >= 0.2 and brightness < 0.6:
        return "Brightness value is " + str(brightness), "Brightness Level is Moderate Brightness"
    else:
        return "Brightness value is " + str(brightness), "Brightness Level is Very Bright"
```

```python
#Tested that the function works
brightness(greenchannel_average)
```

```python
#Defined Red, Green, Blue and Alpha(Transparency) channels
redchannel = dog_img[:,:,0]
greenchannel = dog_img[:,:,1]
bluechannel = dog_img[:,:,2]
alphachannel = dog_img[:,:,3]
```

```python
#Displayed each channel side-by-side for comparison
channels = np.hstack((redchannel, greenchannel, bluechannel,alphachannel))

plt.imshow(channels, cmap = 'gray')
```

```python
#Applied the specific darkness factors

darkened_red = redchannel * 0.8
darkened_green = greenchannel * 0.5
darkened_blue = bluechannel * 0.7
```

```python
#Displayed each darkened channel side-by-side for comparison
darkened_channels = np.hstack((darkened_red, darkened_green, darkened_blue, alphachannel))

plt.imshow(darkened_channels, cmap = 'gray')
```

###### For Red Channel (left-most image), the change is not as evident as for Green and Blue. This is because its darkness factor of 0.8 is high, which means that it still contains about 80% of its brightness. For the Green and Blue channels, it is more evident that they were darkened. This means that the application of darkness factors to the RGB channels are successful. For the next portion, we will be stacking the darkened channel images together along with the unchanged alpha channel to produce the new darkened image of the dog.

```python
#Displayed the new darkened image
darkened_image = np.dstack((darkened_red, darkened_green, darkened_blue, alphachannel))
plt.imshow(darkened_image)
```

```python
#Displayed the corresponding number of channels of the new image

darkened_image.shape[2]
```

###### It can be seen that the new darkened image still has 4 channels which is the number of channels of the original image.

```python
#Assigned height and width variables for the upcoming computation of the centre point

height = darkened_image.shape[0]
width = darkened_image.shape[1]
```

```python
#Got the array for the center point
center_point = darkened_image[height // 2, width // 2]

center_point
```

```python
#Displayed the average pixel value at the center point

ave_pixel_value = np.average(center_point)

print("The average pixel value at the center point is " + str(ave_pixel_value))
```

```python
#Defined the function that will convert the original image to grayscale
def grayscale(image):
    row, column, channel = np.shape(image)
    new_image = np.zeros((row,column))#created a new matrix with the same number of rows and columns, temporarily filled out with zeros
    for i in range(row):
        for j in range(column):
            new_image[i,j] = image[i,j,0:3].min()#got the min value for each array excluding the alpha channel as per the instruction only the minimum value of the first three colour channels are needed
    return new_image
```

```python
#Applied the grayscale function to the original image
grayscale_image = grayscale(dog_img)

#Displayed the original image and the new grayscale version of the image side-by-side

fig, axis = plt.subplots(1,2, figsize = (10,5))
axis[0].imshow(dog_img)
axis[0].set_title("Original Colored Image")

axis[1].imshow(grayscale_image, cmap = 'gray')
axis[1].set_title("New Grayscale Image")

plt.show()
```

```python
#Displayed the pixel value at the center point (height // 2, width // 2) of the resulting grayscale image
center_point_gs = grayscale_image[height // 2, width // 2]

print("The pixel value at the center point is " + str(center_point_gs))
```

```python
#Imported necessary packages
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
#Defined the function
def build_deep_nn(num_hidden, hidden_sizes, dropout_rates, output_size, output_activation):
    nn_model = models.Sequential()
    for i in range(num_hidden):
        if dropout_rates[i] > 0:
            if i == 0:
                nn_model.add(layers.Dense(hidden_sizes[i], activation='relu', input_shape = (100,100,3)))
                nn_model.add(layers.Dropout(dropout_rates[i]))
            else:
                nn_model.add(layers.Dense(hidden_sizes[i], activation='relu'))
                nn_model.add(layers.Dropout(dropout_rates[i]))
        else:
            if i == 0:
                nn_model.add(layers.Dense(hidden_sizes[i], activation='relu', input_shape = (100,100,3)))
            else:
                nn_model.add(layers.Dense(hidden_sizes[i], activation='relu')) 
    nn_model.add(layers.Dense(output_size, activation = output_activation))
    return nn_model
```

```python
#Defined model
model = build_deep_nn(2, (40, 20), (0.0, 0.5), 3, 'sigmoid')
```

```python
#Displayed the model summary
print(model.summary())
```

```python
#Displayed the activation function name of the layers with index=1 and index=3
idx1 = model.layers[1].get_config()['activation']
idx3 = model.layers[3].get_config()['activation']

print("The activation function name of the layer with index=1 is: " + idx1)
print("The activation function name of the layer with index=3 is: " + idx3)
```
